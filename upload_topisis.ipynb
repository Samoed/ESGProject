```python
import pandas as pd
import requests
import json
```

```python
URL = "https://service-index-ai.ethics.hse.ru"
```

# RSPP

```python
df = pd.read_csv("results.csv", index_col=0)
df.head()
```

```python
companies = df["company"].unique().tolist()
```

```python
js_companies = []
for company in companies:
    js_companies.append({"name": company, "company_type_id": 1})
```

```python
res = requests.post(f"{URL}/topsis/company", json={"data": js_companies})
res
```

```python
company_id = {company["name"]: company["id"] for company in res.json()["data"]}
```

```python
for key in company_id.keys():
    if key not in companies:
        print(key)
```

```python
df.fillna(0, inplace=True)
```

```python
topsis_data = []
for _, row in df.iterrows():
    years = row["year"].split(",")
    for year in years:
        topsis_data.append(
            {
                "company_id": company_id[row["company"]],
                "year": int(float(year)),
                "environmental": row["env"],
                "social": row["social"],
                "governance": row["gov"],
                "on_letters": row["on_letters"],
                "all_topics": row["all_letters"],
                "average": row["avg"],
            }
        )
```

```python
res = requests.post(f"{URL}/topsis/topsis", json={"data": topsis_data})
res
```
