```python
topics = pd.read_csv(f"../topic_words_eng.csv.zip")
topics.head()
```

```python
topics["topic"].nunique()
```

```python
set(topics["topic"].unique().tolist()) - set(df["max_topic_cos"].unique().tolist())
```

```python
topics_per_letter = topics.groupby("meta")["topic"].apply(set).to_dict()
```

```python
for name, topics_ in topics_per_letter.items():
    for topic in topics_:
        df.loc[df["max_topic_cos"] == topic, "max_topic_cos"] = name
```

```python
pd.DataFrame(df.groupby(["file_name", "max_topic_cos"])["year"].count()).rename(columns={"year": "count"}).reset_index().to_csv("file_topics_meta.csv", index=False)
```
